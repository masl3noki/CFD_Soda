# TEST 1

$T = 0.25$ time units

```python
import numpy as np
import matplotlib.pyplot as plt

Path = "test1/"

with open(Path + "P1.txt", 'r') as inp:
    lst1 = inp.read().split()
    lst1 = [[float(n) for n in x.split()] for x in lst1]

with open(Path + "U.txt", 'r') as inp:
    lst2 = inp.read().split()
    lst2 = [[float(n) for n in x.split()] for x in lst2]

with open(Path + "E.txt", 'r') as inp:
    lst3 = inp.read().split()
    lst3 = [[float(n) for n in x.split()] for x in lst3]

with open(Path + "Ro.txt", 'r') as inp:
    lst4 = inp.read().split()
    lst4 = [[float(n) for n in x.split()] for x in lst4]


t = np.arange(0, 2, 0.01)
plt.subplot(2,2,1)
plt.title('Co = 0.005, t=0.2, Ro')
plt.plot(t, lst4)

plt.subplot(2,2,2)
plt.title('Co = 0.005, t=0.2, U')
plt.plot(t, lst2)

plt.subplot(2,2,3)
plt.title('Co = 0.005, t=0.2, P')
plt.plot(t, lst1)

plt.subplot(2,2,4)
plt.title('Co = 0.005, t=0.2, E')
plt.plot(t, lst3)
plt.savefig("Godtvd.png")
plt.show()
plt.clf()
```
