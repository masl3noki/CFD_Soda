# Создание графики для теха

Используется библиотека manim: https://docs.manim.community/en/stable/examples.html

```python
import numpy as np
from manim import *
```

```python
%%manim -qh -v WARNING plot_test

class plot_test(Scene):
    def construct(self):
        plane = NumberPlane()
        sec = VMobject(color=RED)
        sec.points = [
            np.array([1, 1, 0]),
            np.array([-3, 2, 0]),
            np.array([0, 0, 0]),
            np.array([3, 0, 0]),
        ]
        self.add(plane, sec)
```

```python
%%manim -qh -v WARNING HelloLaTeX

class HelloLaTeX(Scene):
    def construct(self):

        myTemplate = TexTemplate()
        myTemplate.add_to_preamble(r"\usepackage[T2A]{fontenc}")
        myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}")

        tex = Tex(
            r"$|\pmb{A} - \lambda \pmb{E}| = 	\begin{pmatrix} v - \lambda & \rho & 0\\ 0 & v - \lambda & \frac{1}{\rho}\\ 0 & \gamma p & v - \lambda \end{pmatrix} = (v - \lambda)^3 - \frac{\gamma p}{\rho} (v - \lambda) = 0$", 
            tex_template=myTemplate,
            font_size=30,
        )
        self.add(tex)
```
